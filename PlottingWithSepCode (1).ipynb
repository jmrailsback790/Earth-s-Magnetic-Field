```python
import physplot as phys
import numpy as np
from __future__ import division

theta = np.array([-50, -40, -30, -20, -10, 10, 20, 30, 40, 50])
theta_rad = theta * (3.14 / 180)
tan_theta = np.tan(theta_rad)

I_100N = np.array([-35.2, -25.8, -16.30, -10.49, -4.58, 4.08, 10.34, 16.01, 24.2, 35.7])
I_200N = np.array([-17.52, -13.11, -8.09, -5.2, -2.39, 1.98, 5.22, 9.09, 12.28, 17.83])
err_I_100N = np.array([.2, .2, .2, .2, .2, .2, .2, .2, .2, .2])

#plotfit takes in parameters in this order
#phys.plotfit(x,y,dy,title,'xlabel(units)','ylabel(units)','slope units')  Slope units will not render in LaTeX

phys.plotfit(tan_theta , I_100N, err_I_100N,'100-turn Coil','tan(theta)','Current (mA)','mA')
phys.plotfit(tan_theta , I_200N, err_I_100N,'200-turn Coil','tan(theta)','Current (mA)','mA')

#Determine B
d = 0.218
err_d = 0.001
r = d / 2 
err_r = 0.0005

mew = (12.56637061 * (10 ** -7)) #Units (Tm / A)

N1 = 100
N2 = 200

B1 = (0.029439001706492977 * mew * N1) / (2 * r)
print(B1)
B2 = (0.014873914133106656 * mew * N2) / (2 * r)
print(B2)
slope1 = 0.029439001706492977
slope2 = 0.014873914133106656
err_slope = 0.08741853096010642 / 1000

err_B1 = B1 * np.sqrt((err_r / r)**2 + (err_slope / slope1)**2)
err_B2 = B2 * np.sqrt((err_r / r)**2 + (err_slope / slope2)**2)

print(err_B1, err_B2)
print('In the 100-turn coil, B =', B1, '+/-', err_B1, 'T')
print('In the 200-turn coil, B=', B2, '+/-', err_B2, 'T')
```
